# LOGISTIC REGRESSTION

```python
data1=data.drop("label",axis=1)
```

```python
data1.corr()
```

# Coverting data object to int

```python
data.label =[1 if each == "female" else 0 for each in data.label]
```

```python
data.info()
```

```python
y = data.label.values
x_data = data1
```

```python
x = (x_data - np.min(x_data)) / (np.max(x_data.values))
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
method_name = []
method_scores =[]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
```

```python
log_reg.fit(x_train,y_train)
```

```python
print("Logistic Regression Classification Test Accuracy{}".format(log_reg.score(x_test,y_test)))
```

```python
method_name.append("Logistic Reg")
method_scores.append(log_reg.score(x_test,y_test))
```

```python
y_pred=log_reg.predict(x_test)
```

```python
conf_mat = confusion_matrix(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f, ax=plt.subplots(figsize=(5,5))
sns.heatmap(conf_mat,annot=True,linewidths=0.5,linecolor="red",fmt="0f",ax=ax)
plt.xlabel("Predicted values")
plt.ylabel("True Values")
plt.show()
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_forest = RandomForestClassifier(n_estimators=100,random_state=42)
rand_forest.fit(x_train,y_train)
print("Random Forest Classification:",rand_forest.score(x_test,y_test))
```

```python
method_name.append("Random Forest")
method_scores.append(rand_forest.score(x_test,y_test))
```

```python
# Confustion MATRIX
```

```python

y_pred=rand_forest.predict(x_test)
conf_mat = confusion_matrix(y_test,y_pred)
```

```python
f, ax=plt.subplots(figsize=(5,5))
sns.heatmap(conf_mat,annot=True,linewidths=0.5,linecolor="red",fmt="0f",ax=ax)
plt.xlabel("Predicted values")
plt.ylabel("True Values")
plt.show()
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc=GradientBoostingClassifier(n_estimators=500,learning_rate=0.05,random_state=100,max_features=5)
```

```python
gbc.fit(x_train,y_train)
print("Gradient Forest Classification score :",gbc.score(x_test,y_test))
method_name.append("Gradient Forest")
method_scores.append(gbc.score(x_test,y_test))

## Confustion Matrix

y_pred = gbc.predict(x_test)
conf_mat = confusion_matrix(y_test,y_pred)
f, ax=plt.subplots(figsize=(5,5))
sns.heatmap(conf_mat,annot=True,linewidths=0.5,linecolor="red",fmt="0f",ax=ax)
plt.xlabel("Predicted values")
plt.ylabel("True Values")
plt.show()
```
